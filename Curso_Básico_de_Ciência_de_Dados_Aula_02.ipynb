Em caso de análises mais sérias, perceber que o valor do desvio padrão está muito grande pode indicar a presença de um valor discrepante. Isso pode ter vários significados, como um erro na base de dados, um cálculo mal feito da parte dos administradores, ou até mesmo fraude. 

Essa diferença também pode ser percebida na média, caso ela seja demasiadamente maior que a mediana.
